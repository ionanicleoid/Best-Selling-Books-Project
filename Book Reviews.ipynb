# Best Selling Books Project

This project will use Python data visualisation tools to analyse the Kaggle Amazon Top 50 Best Selling Books.

Dataset available at: https://www.kaggle.com/sootersaalu/amazon-top-50-bestselling-books-2009-2019

Dataset on Amazon's Top 50 bestselling books from 2009 to 2019. Contains 550 books, data has been categorized into fiction and non-fiction using Goodreads.

Attributes

Name - Name of the Book

Author - The author of the Book

User Rating - Amazon User Rating

Reviews- Number of written reviews on amazon

Price- The price of the book (As at 13/10/2020)

Year - The Year(s) it ranked on the bestseller

Genre - Whether fiction or non-fiction


### Load Dataset 

```python
#load pandas
import pandas as pd

#read csv
data= pd.read_csv('bestsellers with categories.csv', delimiter=',',  na_values='?')

#view 5 lines of data
data.head()

#drop null values
data = data.dropna()
```

## Countplot of Genre

```python
#import matplotlib and seaborn
import matplotlib.pyplot as plt
import seaborn as sns

#create a countplot
plt.figure(figsize=(15,8))
sns.countplot(x='Genre', data= data)

#label x ans y axis
plt.xlabel('Genre')
plt.ylabel('Number of Counts')

#show plot
plt.show()
```

We can see that the most common Genre of bestseller books are non-fiction with over 300 counts compared to fiction books at aroung 250. 

### Boxplot per Year

```python
#sort values by Year
data = data.sort_values(by="Year")

#group data by Year
year = data.groupby('Year')
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")
ax = sns.boxplot(x=data["Year"], y=data["Price"])
```

We can see from the boxplot that the average price of books reviewed over 10 years has not hugely differed with the mean of each year remainining under 20. Each year has outliers however 2013 and 2014 have significant outliers with prices over 100. 

### Scatterplot of User Rating and Reviews 

This scatterplot is used to observe relationships between variables of User Rating and Reviews.

```python
import numpy as np
import matplotlib.pyplot as plt


plt.scatter(x=data["User Rating"], y=data["Reviews"], alpha=0.3,
           color="yellow")
plt.show()
```

We can see that the dots are concentrated on the right side of the plot around the value 4.8 on the x-axis, and under 20,000 on the y-axis. We can also see that the spread is wider on the y-axis than on the x-axis.

This suggests that majority of the views of the books are on the higher side. Which would suggest most books are reviewed highly.
